#### Long to wide table format

```python
# filter for no2 data only
no2 = air_quality[air_quality["parameter"] == "no2"]
```

```python
# use 2 measurements (head) for each location (groupby)
no2_subset = no2.sort_index().groupby(["location"]).head(2)
no2_subset
```

```python
no2_subset.pivot(columns = "location", values = "value")
```

```python
no2.head()
```

```python
no2.pivot(columns = "location", values = "value").plot()
```

```python
air_quality.pivot_table(
    values = "value", index = "location", columns = "parameter", aggfunc = "mean"
)
```

```python
air_quality.pivot_table( # 피벗 테이블 생성
    values = "value", # 데이터 값으로 사용할 열 이름
    index = "location", # location 열을 기준으로 행 생성
    columns = "parameter", # parameter 열을 기준으로 열 생성
    aggfunc = "mean", # 피벗 테이블에 적용할 집계 함수
    margins = True, # 피벗 테이블에 마진(Margins)을 추가할지 여부 지정
) # 마진은 각 행과 열에 대한 총계를 나타내는 행과 열을 추가
```

#### Wide to long format

```python
no2_pivoted = no2.pivot(columns = "location", values = "value").reset_index()
```

```python
no2_pivoted.head()
```

```python
no_2 = no2_pivoted.melt(id_vars = "date.utc")
no_2.head()
```

```python
no_2 = no2_pivoted.melt( # 피벗 테이블을 녹여서 긴 형식으로 변환
    id_vars = "date.utc", # 변환된 데이터프레임에서 유지할 열의 이름
    value_vars = ["BETR801", "FR04014", "London Westminster"], # 변환된 데이터프레임에 포함할 값 열의 이름
    value_name = "NO_2", # 변환된 데이터프레임에서 값 열의 이름 지정
    var_name = "id_location", # 값 열의 이름 대신 사용할 변수 열의 이름
)
no_2.head()
```
